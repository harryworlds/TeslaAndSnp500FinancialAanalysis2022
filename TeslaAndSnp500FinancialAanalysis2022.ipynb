### 3. REGRESSION (TSLA & SNP500)

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y)
plt.ylabel("TSLA")
plt.xlabel("SNP500")
plt.show()
```

### 4. PREDICTION

```python
reg = np.polyfit(x,y, deg=1) #prediction x axis is snp500 and y axis is tesla
reg #array([l, m]) #Every l percent of SNP500 will make m percent increase in TSLA
```

### 5. REGRESSION TECHINQUES WITH TREND LINE

```python
plt.figure(figsize=(15,10))
trend = np.polyval(reg, x)
plt.scatter(x,y)
plt.plot(x, trend, "r")
plt.ylabel("TSLA")
plt.xlabel("SNP500")
plt.show()
```

### 7. TSLA TREND LINE

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x="Date", y="Open", data = df_tesla_five, color="red")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("TSLA")
plt.show()
```

```python
plt.figure(figsize=(15,10))
x = np.arange(df_tesla_five["Open"].size)
fit = np.polyfit(x, df_tesla_five["Open"], deg=1)
fit_function = np.poly1d(fit)
sns.lineplot(data=df_tesla_five["Open"], color="red")
sns.lineplot(x=df_tesla_five["Open"].index, y=fit_function(x))
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("TSLA")
plt.show()
```
